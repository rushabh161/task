```python
#Q2
def   analyze_string ( s ) :       
    if  s ==  "" :           
        print ( "Empty string entered" )           
        return       
    print ( "Length:" ,   len ( s ))       
    print ( "Reverse:" ,  s [:: -1 ])      
    
    vowels =  "aeiouAEIOU"      
    count =  0       
    
    for  ch  in  s :           
        if  ch  in  vowels :              
            count +=  1       
            print ( "Vowels count:" ,  count )       
            for  i  in   range ( len ( s )):
                print(f"Char:  { s [ i ]} , Positive:  { i } , Negative:  { i -  len ( s )} ")  
                
                
s =  input ( "Enter a string: " )  
analyze_string ( s )
```

```python
#Q3
def manage_marks():
    marks=[]
    try:
        for i in range(5):
            mark=float(input(f"Enter marks{i+1}:"))
            marks.append(mark)

        print("Average marks:", sum(marks)/5)
        print("HIghest marks:", max(marks))
        print("Lowest marks:", min(marks))
        marks.sort(reverse=True)
        print("Marks in descending order:", marks)
    except ValueError:

        print("Invalid input. Please enter a valid number.")

manage_marks()
```

```python
#Q4
class Student:
    def __init__(self, name, roll_no):

        self.name = name

        self.roll_no = roll_no

        self.marks_list = []


    def add_mark(self, mark):

        try:

            if mark < 0 or mark > 100:

                raise ValueError("Marks must be between 0 and 100")

            self.marks_list.append(mark)

        except ValueError as e:

            print(e)


    def get_average(self):

        if len(self.marks_list) == 0:

            return 0

        return sum(self.marks_list) / len(self.marks_list)


    def display_info(self):

        print("\nStudent Details")

        print("Name:", self.name)

        print("Roll No:", self.roll_no)

        print("Marks:", self.marks_list)

        print("Average:", self.get_average())


name = input("Enter Name: ")

roll = input("Enter Roll Number: ")

student = Student(name, roll)

for i in range(5):

    while True:

        try:

            mark = float(input(f"Enter Mark {i+1}: "))

            student.add_mark(mark)

            break

        except ValueError:

            print("Invalid Input!")

student.display_info()
```
